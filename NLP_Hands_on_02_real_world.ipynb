##Exploring the features of NLTK  
The first steps in the tutorial load and view a piece of sample text, and import the libraries of NLTK that will be in use.

```python
# Open the text file for processing

#To get the text file used :
!wget https://raw.githubusercontent.com/robertwesterman/NLPtutorials/main/USNews20200624PoliceMisconductCases.txt

#Open the text file :
text_file = open("USNews20200624PoliceMisconductCases.txt")

#Read the data :
text = text_file.read()

#Datatype of the data read :
print (type(text))
print("\n")

#Print the text :
print(text)
print("\n")
#Length of the text :
print (len(text))
```

```python
#Import required libraries :
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize


import nltk
nltk.download("popular")
```

##Tokenization  
Converting the items between the spaces in the text file to words and sentences.

```python
#Tokenize the text by sentences :
sentences = sent_tokenize(text)

#How many sentences are there? :
print (len(sentences))

#Print the sentences :
#print(sentences)
sentences
```

```python
#Tokenize the text with words :
words = word_tokenize(text)

#How many words are there? :
print (len(words))
print("\n")

#Print words :
print (words)
```

```python
# Frequency distribution of words (and punctuation) in the text

#Import required libraries :
from nltk.probability import FreqDist

#Find the frequency :
fdist = FreqDist(words)

#Print 10 most common words :
fdist.most_common(10)
```

```python
#Plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

```python
# Removing punctuation by selecting for alphanumeric characters

#Empty list to store words:
words_no_punc = []

#Removing punctuation marks :
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the words without punctution marks :
print (words_no_punc)

print ("\n")

#Length :
print (len(words_no_punc))
```

```python
# Word frequency without punctuation

#Frequency distribution :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

```python
# Word frequency graphed without punctuation

fdist.plot(10)
```

##Stopwords  
The frequent words list, above, shows many conjunctions and articles that contribute little to the meaning of the sentences. These and words like them are called "stopwords", and are removed in the following steps.

```python
# Import list of stopwords

from nltk.corpus import stopwords

#List of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

```python
# Collect words from sample text that are not stopwords

#Empty list to store clean words :
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)
        
print(clean_words)
print("\n")
print(len(clean_words))
```

```python
# Frequency of text without stopwords


fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
#Plot the most common words on grpah:

fdist.plot(10)
```

##Stemming  
Stemming is a method to convert the various forms of a word to single common root. This more directly gets at the common underlying meaning of the different forms that might appear in the raw text.  
NLTK incorporate's multiple stemming libraries, two of which are demonstrated here.

```python
# Porter Stemmer example

#Import stemming library :
from nltk.stem import PorterStemmer

porter = PorterStemmer()

#Word-list for stemming :
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(porter.stem(w))
```

```python
# Snowball Stemmer example

#Import stemming library :
from nltk.stem import SnowballStemmer

snowball = SnowballStemmer("english")

#Word-list for stemming :
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(snowball.stem(w))
```

Snowball Stemmer supports multiple languages:

```python
#Print languages supported :
SnowballStemmer.languages
```

The tutorial then overviews the pros and cons of six stemmers, and concludes that Porter Stemmer produces the best output, and has the lowest error rate.

##Lemmatization  
This process attempts to convert words to a more basic form with the same essential meaning, just as stemming, but arrives at a dictionary word rather than a truncated stem. Lemmatization considers the context of a word, and produces results that are more accurate, but much slower than, stemming.

```python
# Lemmatization

from nltk import WordNetLemmatizer

lemma = WordNetLemmatizer()
word_list = ["am","is","are","was","were"]

for w in word_list:
    print(lemma.lemmatize(w ,pos="v"))
```

```python
# Part of speech can be forced on the word interpretation, causing it to be reduced to the shortest format of the word with that PoS definition

lemmatizer = WordNetLemmatizer()
print(lemmatizer.lemmatize('studying', pos="v"))
print(lemmatizer.lemmatize('studying', pos="n"))
print(lemmatizer.lemmatize('studying', pos="a"))
print(lemmatizer.lemmatize('studying', pos="r"))
```

##Part of Speech Tagging  
To get from categorizing and simplifying text to syntactic and semantic analysis requires identifying each token's part of speech. A simple example is separating the two applications of the word *can* in the following sentence: **"Can you help me with the can?"**  

The tutorial then lists 35 parts of speech, their NLTK POS tags, a few words in that category, and use in a sample sentence.

##Chunking  
Paragraphs, which are easy for people to read, get broken down into tokens, which are easy for computers to use. The preceding steps eliminated non-contributing words (stopwords,) simplified words to a root form, and identified parts of speech. The next step involves assessing those (now) data-friendly elements to extract meaningful phrases. That process of considering the meaning of words in context with the meaning of other words in the same and nearby sentences is called **chunking**.  
Here the tutorial briefly mentions 5 categories of phrases and that they fit together in structured ways to form sentences. The relevance to this example is that the module used to parse the phrase grammar (do the "chunking") takes direction as to what types of phrases it's looking for. The following example takes the sentence from the previous example and attempts to identify noun phrases. The code defines the noun phrase as a determiner (optional) followed by adjectives (zero to many) and a noun.

```python
# Extracting Noun Phrase from text

# ? - optional character
# * - 0 or more repetations
grammar = "NP : {<DT>?<JJ>*<NN>} "
#Creating a parser :
parser = nltk.RegexpParser(grammar)

#Parsing text :
output = parser.parse(tagged_words)
print (output)
```

```python
# Visualizing Noun Phrase from text

# ? - optional character
# * - 0 or more repetations
grammar = "NP : {<DT>?<JJ>*<NN>} "
import matplotlib.pyplot as plt
#Creating a parser :
parser = nltk.RegexpParser(grammar)

#Parsing text :
output = parser.parse(tagged_words)

# To visualize
output.draw()
```

##Chinking  
This step excludes elements of text from the chunks. This can lead to better output with complex extractions.  
The following example removes adjectives from noun phrases. As opposed to chunking, the grammar is defined within inverted (reversed) curly braces:  
`} write chinking grammar here {`

```python
# Chinking example
# * - 0 or more repetations
# + - 1 or more repetations

#Here we are taking the whole string and then
#excluding adjectives from that chunk.

grammar = r""" NP: {<.*>+} 
               }<JJ>+{"""

#Creating parser :
parser = nltk.RegexpParser(grammar)

#parsing string :
output = parser.parse(tagged_words)
print(output)
```

##Named Entity Recognition (NER)  
These items are proper nouns, and numerical nouns such as dates, times, currency amounts, and other numbers. An argument in the `ne_chunk()` command, `binary`, sets the level of detail in the output.  
###1. binary = True  
This method identifies named entities, but no further detail.  
###2. binary = False  
This method identifies the subcategory of named entity.


```python
# NER example 1, binary=True

#Sentence for NER :
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words)

#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=True)

print(N_E_R)
```

```python
# NER example 2, binary=False

#Sentence for NER :
sentence = "Mr. Smith made a deal on a beach of Switzerland near WHO."

#Tokenizing words :
tokenized_words = word_tokenize(sentence)

#PoS tagging :
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words)

#Named Entity Recognition :
N_E_R = nltk.ne_chunk(tagged_words,binary=False)
print(N_E_R)
```

##Meanings of Words: WordNet  
Part of the NLTK corpus, Wordnet is a database of definitions, synonyms, and antonyms in English. The examples below show:  
###1. Number of definitions available for a word  
###2. The meanings of each definition  
###3. All details for a word  
###4. All details for all meanings of a word  
###5. Hypernyms (more abstract terms for a word)  
###6. Hyponyms (more specific terms, or subcategories, for a word)  
###7. Synonyms and Antonyms  
###8. Similarity between words

```python
# Number of definitions available for a word

#Import wordnet :
from nltk.corpus import wordnet

for words in wordnet.synsets("Fun"): 
    print(words)  
```

```python
# The meanings of each definition

for words in wordnet.synsets("Fun"): 
    for lemma in words.lemmas(): 
        print(lemma)
    print("\n")
```

```python
# All details for a word

word = wordnet.synsets("Play")[0]

#Checking name :
print(word.name())

#Checking definition :
print(word.definition())

#Checking examples:
print(word.examples())
```

```python
# All details for all meanings of a word

for words in wordnet.synsets("Fun"): 
    print(words.name())
    print(words.definition())
    print(words.examples())
    
    for lemma in words.lemmas(): 
        print(lemma)
    print("\n")
```

```python
# Hypernyms (more abstract terms for a word)

word = wordnet.synsets("Play")[0]

#Find more abstract term :
print(word.hypernyms())
```

```python
# Hyponyms (more specific terms, or subcategories, for a word)

word = wordnet.synsets("Play")[0]

#Find more specific term :
word.hyponyms()
```

```python
# Synonyms and Antonyms

#Empty lists to store synonyms/antonynms : 
synonyms = []
antonyms = []

for words in wordnet.synsets('New'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())
            
#Print lists :
print(synonyms)
print("\n")
print(antonyms)
```

```python
# Similarity between words

word1 = wordnet.synsets("ship","n")[0]

word2 = wordnet.synsets("boat","n")[0] 

#Check similarity :
print(word1.wup_similarity(word2))
```

```python
# Similarity between words, example 2

word1 = wordnet.synsets("ship","n")[0]

word2 = wordnet.synsets("bike","n")[0] 

#Check similarity :
print(word1.wup_similarity(word2)) 
```

##Bag of Words  
This is a method for extracting features from row text to be used for machine learning models. The term referrs to the fact that identified words and their frequency are used, but order of the words is irrelevant.  
The Bag of Words structure involves 5 steps:  


1.   Raw Text - the original text to be analysed
2.   Clean Text - remove stopwords & punctuation
3.   Tokenize - convert words to tokens
4.   Building Vocab - arrive at a list of words
5.   Generate Vectors - word frequency by sentence  
  
The result is a table of words (columns) and sentences (rows), with the interior cells showing the number of times each word appeared in each sentence. This type of output is useful for information retrieval and classification of documents. However, it loses semantic meaning and large documents result in huge vector arrays.  
This example shows first a dictionary of each word in the order it occurs in the raw text, along with a number indicating its column in the table. Second is a list of the words, in the order of their table columns. Finally is the table itself, a list of lists.



```python
# Bag of Words example

#Import required libraries :
from sklearn.feature_extraction.text import CountVectorizer

#Text for analysis :
sentences = ["Jim and Pam travelled by the bus:",
             "The train was late",
             "The flight was full.Travelling by flight is expensive"]

#Create an object :
cv = CountVectorizer()

#Generating output for Bag of Words :
B_O_W = cv.fit_transform(sentences).toarray()

#Total words with their index in model :
print(cv.vocabulary_)
print("\n")

#Features :
print(cv.get_feature_names())
print("\n")

#Show the output :
print(B_O_W)
```

##TF-IDF  
Term Frequency-Inverse Document Frequency is a scoring method used in information retrieval and summarization. The score shows the relevance or importance of a term in a given document.  
The scoring balances how frequent a word is in the target document against how frequently the word appears in documents generally. The terms have mathematical definitions:  
**Term Frequency**  
$TF=\frac{\text{Frequency of the word in the sentence}}{\text{Total number of words in the sentence}}$  
<br />
**Inverse Document Frequency**  
$IDF=\frac{\text{Total number of sentences}}{\text{Number of sentences containing that word}}$  
<br />
The final score is the product of TF and IDF.  
One way to get more useable scores for large documents is to recalculate using the log of the IDF value, then add one, before multiplying by TD. This is the method used by the **Sklearn** function in the following example.

```python
#Import required libraries :
from sklearn.feature_extraction.text import TfidfVectorizer

#Sentences for analysis :
sentences = ['This is the first document','This document is the second document']

#Create an object :
vectorizer = TfidfVectorizer(norm = None)

#Generating output for TF_IDF :
X = vectorizer.fit_transform(sentences).toarray()

#Total words with their index in model :
print(vectorizer.vocabulary_)
print("\n")

#Features :
print(vectorizer.get_feature_names())
print("\n")

#Show the output :
print(X)
```
